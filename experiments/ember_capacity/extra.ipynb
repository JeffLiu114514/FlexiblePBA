```python
import itertools
import numpy as np

def get_permutations(lst):
    return list(itertools.permutations(lst))

gray_coding = \
{
    4: ["00", "01", "11", "10"],
    8: ["000", "001", "011", "010", "110", "111", "101", "100"],
    16: ["0000", "0001", "0011", "0010", "0110", "0111", "0101", "0100", "1100", "1101", "1111", "1110", "1010", "1011", "1001", "1000"]
}

def str_diff(s1, s2):
    assert len(s1) == len(s2)
    diff = 0
    for i in range(0, len(s1)):
        if s1[i] != s2[i]:
            diff += 1
    assert diff <= len(s1)
    return diff

def init_dist(coding):
    dist_4 = np.zeros((4, 4))
    dist_8 = np.zeros((8, 8))
    dist_16 = np.zeros((16, 16))

    for i in range(0, 4):
        for j in range(0, 4):
            dist_4[i][j] = str_diff(coding[4][i], coding[4][j])
    for i in range(0, 8):
        for j in range(0, 8):
            dist_8[i][j] = str_diff(coding[8][i], coding[8][j])
    for i in range(0, 16):
        for j in range(0, 16):
            dist_16[i][j] = str_diff(coding[16][i], coding[16][j])
    return dist_4, dist_8, dist_16
    # pprint.pprint(dist_4)
    # pprint.pprint(dist_8)
```
